loading our libraries


```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

loading the preprocessed data

```python
train_df = pd.read_csv('content/preprocessed_train.csv').drop(columns= ['Unnamed: 0'] ) #.query('building_id==1 & meter==1')
# setting the timestamp to the right form
train_df["timestamp"] = pd.to_datetime(train_df["timestamp"])
```

```python
# printing our columns
train_df.head()
```

## taking only building_id 122 , meter 0, primary_use=0

```python
df= train_df.query('building_id==122 & meter==0')
```

```python
df.head()
```

```python
df.meter_reading.plot()
```

```python
df.drop( columns=['precip_depth_1_hr','cloud_coverage'], inplace=True)
df.set_index('timestamp', inplace=True)
```

### Now we have our data 

```python
df.meter_reading.plot()
```

```python
df.rolling('D').mean().meter_reading.plot()
```

```python
df.resample('D').mean().meter_reading.plot()
```

## we will work with resampled data by Day

```python
resampled_data= df.resample('D').mean().drop(columns=['hour', 'site_id', 'square_feet', 'building_id','primary_use', 'meter'])
```

```python
resampled_data
```

```python
resampled_data.describe()
```

# Let's plot it in a nice way before we save it for modeling

```python
import seaborn as sns
```

```python
fig, axs= plt.subplots(6,1, figsize=(15,15))
for i,col in enumerate(resampled_data.columns[:6]):

    sns.lineplot(resampled_data.index ,
             col, 
             data=resampled_data,
             ax= axs[i] )
```

## we can plot the correlation

```python
sns.heatmap(resampled_data.drop(columns=['day','month']).corr() )
```

```python
resampled_data.drop(columns=['day','month']).corr()
```

we can see that there's a slight correlation between meter_reading and temperature

# let's save this data to model it.

```python
resampled_data.to_csv('content/resampled_by_day_id_122.csv')
```
